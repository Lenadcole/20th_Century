### 1. Importing libraries

```python
# Import libraries

from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

### 2. Tokenisation

```python
myfile = open('20th_century_Wiki_Lena_Cole.txt', encoding='utf-8')
```

```python
# Import text file

with open('20th_century_Wiki_Lena_Cole.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '')
```

```python
# Sentence tokenisation

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenisation

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Create frequency distribution

from nltk.probability import FreqDist
dist_words =  FreqDist(tokenized_word)
print(dist_words)
```

```python
dist_words.most_common(10)
```

```python
# Create dataframe

df_dist_words = pd.DataFrame(tokenized_word)
df_dist_words.columns = ['Words']

df_dist_words.head()
```

### 3. Top 10 most common words

```python
# Find the top 10 most common words

top_10_most_common_words = (df_dist_words['Words'].value_counts()).iloc[:10]
```

```python
# Create bar chart to plot the top 10 most common words

top_10_most_common_words.plot(kind='bar')
```

### 4. Removing stopwords

```python
# Define stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
# Remove stopwords in words

filtered_words = [] # This creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
print(filtered_words)
```

```python
# Substitute punctuation marks with a space

sans_punc = re.sub("[^a-zA-Z]",
                   " ",
                   str(filtered_words))
```

```python
sans_punc
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Create new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(10,cumulative = False)
plt.show()
```

```python
# Create dataframe

df_dist_words_filter_2 = pd.DataFrame(tokenized_word_2)
df_dist_words_filter_2.columns = ['Words']

df_dist_words_filter_2.head()
```

```python
# Find the top 10 most common words

top_10_post = (df_dist_words_filter_2['Words'].value_counts()).iloc[:10]
```

```python
# Create bar chart to plot the top 10 most common words post filtering stop words and punctuation

top_10_post.plot(kind='bar')
```

The newly filtered list features far more actual words than the first one, which was mainly populated with punctuation marks. 'The' remains the most frequent word, so it will be important to filter that out also. 's' and 'th' also form part of this list - the 's' because it was left as a singular letter after apostrophes were removed and 'th' because it follows many numbers. They will have to be filtered out.
Interestingly, both 'War' and 'war' feature on this list - We did not cover this issue in this exercise, but it would be useful to filter words regardless of case, if possible.

```python
# Add new words to stopwords list

new_stopwords = ['The', 's', 'th', 'S']
```

```python
filtered = []
for element in tokenized_word_2:
    if element not in new_stopwords:
        filtered.append(element)
```

```python
filtered
```

```python
# Create TextBlob object

text = TextBlob(str(filtered))
```

```python
text
```

### 5. Creating tags list using TextBlob

```python
# Create list of tags 

tags_list = text.tags
```

```python
tags_list
```

### 6. Top 10 POS tags

```python
# Create list of top 10 POS tags for words that appear on the webpage 

df_text = pd.DataFrame(tags_list)
df_text.columns = ['Words', 'Word type']
df_t = df_text.groupby('Word type').count().reset_index()
top10 = df_t.nlargest(10, 'Words')
```

```python
top10
```

### 7. Plotting the Top 10 POS tags

```python
# Create bar chart of top 10 POS tags

plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:green", 10):
    sns.barplot(x = "Words", y = "Word type",
    saturation = 0.9, data = top10).set_title("Top 10 word types used in 'Key events of the 20th Century' Wikipedia page")
```

### 8. Top 15 nouns, verbs and adjectives

```python
# Find the 15 most common nouns in the article

df_nouns = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP") | (df_text['Word type'] == "NNPS")]
df_nouns.columns = ["Word", "Occurrences"]
x = df_nouns.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top15_nouns= y.nlargest(15, 'Occurrences') 
```

```python
top15_nouns
```

```python
# Create bar chart of top 15 nouns in article

plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:red", 15):
    sns.barplot(x = "Word", y = "Occurrences",
    saturation = 0.9, data = top15_nouns).set_title("Top 15 nouns used in 'Key events of the 20th Century' Wikipedia page")
```

I was surprised by how often the word 'December' appeared in the article. Did more than 200 important events really take place at the same time of year? A look back at the article confirmed that this was not the case! A review of the references section showed that many supporting documents were retrieved in December 2018, which skewed the data. The same could be said for the word 'war'. It appears twice in the nouns list, due to the fact that one instance uses an uppercase 'W' and one does not. If they were treated as the same word (which they are!), they would have a higher number of occurences than 'December'. The rest of the list is taken up by words that are in regards to wars, webpage features ('www', 'edit') or to do with books mentioned in the references ('ISBN' numbers).

```python
# Find the 15 most common verbs in the article

df_verbs = df_text[(df_text['Word type'] == "VB") | (df_text['Word type'] == "VBD") | (df_text['Word type'] == "VBG") | (df_text['Word type'] == "VBN") | (df_text['Word type'] == "VBP") | (df_text['Word type'] == "VBZ")]
df_verbs.columns = ["Word", "Occurrences"]
x = df_verbs.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top15_verbs = y.nlargest(15, 'Occurrences') 
```

```python
top15_verbs
```

```python
# Create bar chart of top 15 verbs in article

plt.figure(figsize = (15, 5))
with sns.dark_palette("xkcd:orange", 15):
    sns.barplot(x = "Word", y = "Occurrences",
    saturation = 0.9, data = top15_verbs).set_title("Top 15 verbs used in 'Key events of the 20th Century' Wikipedia page")
```

As we noticed when looking at the top 15 nouns, the highest number of mentions by far is for a word used in the references section of the webpage - 'Retrieved'. The United States is split across two words with a similar number of mentions and only features in the verbs list because those two words taken on their own are verbs, but are not actually used that way in real life! This shows that tokenisation is really a blunt instrument and context is needed to make it a useful tool. 'Allied' has the same issue - yes, it is a passive verb (To be allied with something), but in the context of the article is the collective name for a set of countries fighting for a common cause.
For those words that are correctly being identified as verbs, many of them have negative connotations - controlled, forced, surrendered - and others are likely to have been used in a military context - led and launched, for example. 'Publishing' and 'Archived' are also words found most often in the references section, similar to some of the most used nouns.

```python
# Find the 15 most common adjectives in the article

df_adjs = df_text[(df_text['Word type'] == "JJ") | (df_text['Word type'] == "JJR") | (df_text['Word type'] == "JJS") | (df_text['Word type'] == "VBN") | (df_text['Word type'] == "VBP") | (df_text['Word type'] == "VBZ")]
df_adjs.columns = ["Word", "Occurrences"]
x = df_adjs.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top15_adjectives = y.nlargest(15, 'Occurrences') 
```

```python
top15_adjectives
```

```python
# Create bar chart of top 15 adjectives in article

plt.figure(figsize = (15, 5))
with sns.dark_palette("xkcd:pink", 15):
    sns.barplot(x = "Word", y = "Occurrences",
    saturation = 0.9, data = top15_adjectives).set_title("Top 15 adjectives used in 'Key events of the 20th Century' Wikipedia page")
```

Several of the most mentioned adjectives belong to countries or areas, which is unsurprising when refering to key events across the world. West has been taken to mean an adjective, although in context it refers to 'The West', a noun that generally means the countries of Europe, North and South America, and Australasia. 'States' has once again been taken out of context, as the tokenisation removed it from its partner word to create the country name 'United States'. The remaining adjectives clearly refer to politics - 'interest', 'power' and 'political' being the most obvious. 'Guardian' has had its leading 'The' removed by stopwords and so it would be easy to miss that it refers to the UK newspaper and that the mentions come from the references section.

### 9. Creating countries dataframe

```python
# Convert list of work tokens into string

listToStr = ' '.join([str(elem) for elem in filtered])

print (listToStr)
```

```python
# Create a count of countries included in the article

all_counts = Counter(re.sub(r'\W+', ' ', listToStr).split())
```

```python
all_counts
```

```python
# Import countries data

countries = pd.read_csv('countries_list_20th_century_1.5.csv', index_col = 0)
```

```python
countries.head()
```

```python
# Replace names with aliases

countries['country_name'] = countries['country_name'].replace('United States', 'States')
countries['country_name'] = countries['country_name'].replace('United Kingdom', 'Britain')
```

```python
countries
```

```python
countries_list = countries['country_name'].to_list()
```

```python
countries_list
```

```python
dict_of_counts = {d : all_counts[d] for d in countries_list}
```

```python
dict_of_counts
```

```python
# Search for the names from the list in the dictionary

dct = {v:[k] for v,k in dict_of_counts.items()}
df_countries = pd.DataFrame(dct)
```

```python
df_countries = df_countries.T
```

### 10. Creating bar plot showing the frequency of the number of times countries are mentioned

```python
# Create bar chart of 20 most mentioned countries

plt.figure(figsize=(10, 5))
with sns.dark_palette('#55A868', 20):
    sns.barplot(x = 'Times mentioned', y = 'Country',
        data = Top_20_most_mentioned_countries)
plt.title('Key events of the 20th century - 20 most frequently mentioned countries')
plt.xlabel('Times mentioned')
plt.ylabel('Country')
plt.show()
```

Instead of plotting all 209 countries, I chose to only plot the 20 most mentioned countries in the article. It is striking just how much the two World Wars dominate the list - Germany is by far the most mentioned country, with the other big players in those wars featuring heavily. If I had not renamed the United States and United Kingdom in the way that I did, neither would have been part of the list - the United States due to the results of tokenisation, and the United Kingdom due to naming conventions of that country (although it would have suffered from the results of tokenisation even if it had been called United Kingdom in the article instead of Britain). If I had wanted to be more thourough, I would have renamed more countries in the list, as there are several more with names of more than one word.
Of the other names on the list, all of them are mentioned with regards to the two World Wars (with the exception of Vietnam, which was mostly mentioned with regards to the Vietnam War).
