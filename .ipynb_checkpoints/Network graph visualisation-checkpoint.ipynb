### 1. Importing libraries and data

```python
# Import libraries

import pandas as pd
import numpy as np
import networkx as nx
import os
import matplotlib.pyplot as plt
from pyvis.network import Network
import seaborn as sns
import pyvis
```

```python
# Import relationship data

df_relationship = pd.read_csv('countries_relationships.csv', index_col = 0)
```

```python
df_relationship.head()
```

```python
df_relationship.shape
```

### 2. Creating network object from dataframe

```python
# Create netowrk object

G = nx.from_pandas_edgelist(df_relationship,
                            source = 'source',
                            target = 'target',
                            edge_attr = 'value',
                            create_using = nx.Graph())
```

### 3. Creating a static network graph

```python
# Create static network graph using NetworkX

h = plt.figure(figsize = (12,12))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, with_labels = True, node_color = 'red', edge_cmap = plt.cm.Blues, pos = pos)
plt.show()
```

```python
h.savefig('networkX_plt_countires.png')
```

### 4. Building interactive network graph

```python
# Build interactive network graph from network object

# Define net object

net = Network()
net = Network(notebook = True, width='900px', height='800px', bgcolor='#222222', font_color='white')

# Define nodes from the G object

node_degree = dict(G.degree)

# Set up node size attributes

nx.set_node_attributes(G, node_degree, 'size')
net.from_nx(G)
net.show_buttons(filter_ = True)
net.repulsion()
net.show('countries.html')
```

### 5. Using Leiden algorithm for community detection

```python
# Import Leiden algorithm

from cdlib import algorithms
coms = algorithms.leiden(G)
```

```python
type(coms)
```

```python
new_com = coms.to_node_community_map()
```

```python
new_com
```

```python
# Put output from new_com into dictionary

dict_com = {k:v[0] for k, v in new_com.items()}
```

```python
# Rerun network graph with communities as an argument

nx.set_node_attributes(G, dict_com, 'group')
com_net = Network(notebook = True, width='1300px', height='700px', bgcolor='#222222', font_color='white')
com_net.from_nx(G)
com_net.show_buttons(filter_=True)
com_net.repulsion()
com_net.show('countries_communities_leiden.html')
```

The partitions make a lot of sense when viewed through the lens of 20th century history. Each community is roughly split into continents or sections of continents. The dark blue community represents Africa, the red Asia, and the gold, the Middle East. Europe is split into several communities: medium blue for North and Western Europe, purple for the countries bordering Germany to the north, yellow for those to its east and green for those to its south east. The pink community is slightly different in that it links the North American countries with Vietnam and so appears to have more a focus on the Vietnam war than the World Wars.

The edges emanating from Germany are thickest when connecting to some of the bigger countries it occupied or attacked during WWII, as well as to its allies in that war. The same could be said of Japan, although in that instance it was Russia whose declaration of war was significant in the decision for Japan to surrender.

### 6. Calculate centrality measures

```python
# Calculate degree centrality

degree_dict = nx.degree_centrality(G)
degree_dict
```

```python
# Create dataframe

df_degree = pd.DataFrame(degree_dict.items(), columns=['Countries','Centrality'])
```

```python
df_degree.sort_values(by = ['Centrality'], ascending=False, inplace=True)
```

```python
# Plot degree centrality

plt.figure(figsize = (20,20))
with sns.dark_palette('#79c', 65):
    sns.barplot(x = 'Centrality', y = 'Countries', 
    saturation = 0.9, data = df_degree).set_title('Countries - degree centrality')
```

Unsurprisingly, Germany has the highest degree of centrality, as it has the most connections to other countries in regards to the history of the 20th century. The next 5 countries are those who had the thickest edges when we ran the Leiden algorithm. At the other end of the list we find countries who had the fewest and thinnest edges in the network graph created after running the Leiden algorithm, among them Spain and Lebanon.

```python
# Calculate closeness centrality

closeness_dict = nx.closeness_centrality(G)
closeness_dict
```

```python
# Create dataframe

df_closeness = pd.DataFrame(closeness_dict.items(), columns=['Countries','Centrality'])
```

```python
df_closeness.sort_values(by = ['Centrality'], ascending=False, inplace = True)
```

```python
# Plot closeness centrality

plt.figure(figsize = (20,20))
with sns.dark_palette('green', 65):
    sns.barplot(x = 'Centrality', y = 'Countries', 
    saturation = 0.9, data = df_closeness).set_title('Countries - closeness centrality')
```

Once again, Germany is top of the list by quite a measure. However, the rest of the countries are relatively near each other on the closeness scale, due to their relationship with Germany during WWII. Persia is noticeably less connected, due to no longer existing as a country during WWII.

```python
# Calculate betweenness centrality

betweenness_dict = nx.betweenness_centrality(G)
betweenness_dict
```

```python
# Create dataframe

df_betweenness = pd.DataFrame(betweenness_dict.items(), columns=['Countries', 'Centrality'])
```

```python
df_betweenness.sort_values(by = ['Centrality'], ascending=False, inplace = True)
```

```python
# Plot betweenness centrality

plt.figure(figsize = (20,20))
with sns.dark_palette('red', 65):
    sns.barplot(x = 'Centrality', y = 'Countries', 
    saturation = 0.9, data = df_betweenness).set_title('Countries - betweenness centrality')
```

The betweenness measure output reconfirms what we saw when looking at the degree and closeness centrality scores - that Germany was the most influential country of the 20th century. Although Japan comes next on the list, its score is half that of Germany, and other countries are not much further behind.

```python
# Save centrality measures

nx.set_node_attributes(G, degree_dict, 'degree_centrality')
nx.set_node_attributes(G, closeness_dict, 'closeness_centrality')
nx.set_node_attributes(G, betweenness_dict, 'betweenness_centrality')
```
